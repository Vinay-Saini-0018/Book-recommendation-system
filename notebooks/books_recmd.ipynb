# Recommendation books

```python
x = books_with_ratings.groupby('User-ID').count()['Book-Title']>=200
```

```python
padhe_likhe_users = x[x].index
```

```python
#users who read more than 200 books
filtered_users = books_with_ratings[books_with_ratings['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_users.shape
```

```python
y = filtered_users.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_books = y[y].index
```

```python
filtered_books = filtered_users[filtered_users['Book-Title'].isin(famous_books)]
```

```python
filtered_books.head()
```

```python
filtered_books.shape
```

```python
pivot_table = filtered_books.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivot_table
```

```python
pivot_table.fillna(0,inplace=True)
```

```python
pivot_table.head()
```

```python
pivot_table.shape
```

```python
#finding the similarity between the books
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pivot_table)
```

```python
similarity_score[0]
```

```python
similarity_score.shape
```

```python
#function for recommended books
def recommend(book_name):
    #fetching index
    index = np.where(pivot_table.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[0])),key=lambda x:x[1],reverse=True)[1:6]

    for i in similar_books:
        print(pivot_table.index[i[0]])   #i[0] is the index of similar books
    
```

```python
recommend('1984')
```

# final function of recommendation books

```python
def recommendall(book_name):
    #fetching index
    index = np.where(pivot_table.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

    data = []
    for i in similar_books:
        item = []
        temp_dt = books[books['Book-Title']==pivot_table.index[i[0]]]
        item.extend(list(temp_dt.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_dt.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_dt.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
recommendall("Midnight in the Garden of Good and Evil: A Savannah Story")
```
